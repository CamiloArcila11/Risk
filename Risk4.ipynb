```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors

from random import randint
counts=dict()
atk =list()
dfnd =list()
contatk, ganatk, perdatk = 0,0,0
y=10000 #número de iteraciones
w=10 #número de atacantes y defensores en la tabla final
h=w

Matrix = [[0 for x in range(w)] for y in range(h)]
for i in range(w):
    contatk, ganatk, perdatk = 0,0,0
    na = i+2
    for j in range(h):
        contatk, ganatk, perdatk = 0,0,0
        nd = j+1
        for i in range(y):
            numatk =na
            numdfnd =nd
            while True:
                if numatk >3 and numdfnd>=2:
                    contint=0
                    atk= sorted((randint(1,6),randint(1,6),randint(1,6)), reverse=True)
                    dfnd=sorted((randint(1,6),randint(1,6)), reverse=True)
                    if atk[0]>dfnd[0]:numdfnd+=-1
                    else: numatk+=-1
                    if atk[1]>dfnd[1]:numdfnd+=-1
                    else: numatk+=-1
                    continue
                if numatk >3 and numdfnd ==1:
                    contint=0
                    atk= sorted((randint(1,6),randint(1,6),randint(1,6)), reverse=True)
                    dfnd=randint(1,6)
                    if atk[0]>dfnd:numdfnd+=-1
                    else: numatk+=-1
                    continue
                if numatk == 3 and numdfnd>=2:
                    contint=0
                    atk= sorted((randint(1,6),randint(1,6)), reverse=True)
                    dfnd=sorted((randint(1,6),randint(1,6)), reverse=True)
                    if atk[0]>dfnd[0]:numdfnd+=-1
                    else: numatk+=-1
                    if atk[1]>dfnd[1]:numdfnd+=-1
                    else: numatk+=-1
                    continue
                if numatk == 2 and numdfnd>=2:
                    contint=0
                    atk= randint(1,6)
                    dfnd=sorted((randint(1,6),randint(1,6)), reverse=True)
                    if atk>dfnd[0]:numdfnd+=-1
                    else: numatk+=-1
                    continue
                if numatk == 3 and numdfnd ==1:
                    contint=0
                    atk= sorted((randint(1,6),randint(1,6)), reverse=True)
                    dfnd=randint(1,6)
                    if atk[0]>dfnd:numdfnd+=-1
                    else: numatk+=-1
                    continue
                if numatk == 2 and numdfnd==1:
                    contint=0
                    atk= randint(1,6)
                    dfnd=randint(1,6)
                    if atk>dfnd:numdfnd+=-1
                    else: numatk+=-1
                    continue
                if numatk==1:
                    perdatk+= 1
                    break
                if numdfnd==0:
                    ganatk+= 1
                    break

        Matrix[nd-1][na-2] = ganatk/y

df=pd.DataFrame(Matrix)
df.index = np.arange(1,len(df)+1)
df.columns = np.arange(1,len(df)+1)

def background_gradient(s, m, M, cmap='YlGn', low=0, high=0):
    rng = M - m
    norm = colors.Normalize(m - (rng * low),
                            M + (rng * high))
    normed = norm(s.values)
    c = [colors.rgb2hex(x) for x in plt.cm.get_cmap(cmap)(normed)]
    return ['background-color: %s' % color for color in c]

df.index.name = 'Defnd\Atck'
df.style.apply(background_gradient,
               cmap='YlGn',
               m=df.min().min(),
               M=df.max().max(),
               low=0,
               high=0.2).format('{:.0%}').set_caption("Probabilidades de ganar atacando en Risk").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-size', '18px')
    ]
}])

```
